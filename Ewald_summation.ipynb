# Ewald summation

- special case of Possion summation formula
- replacing the summation of interacting energies in real space with an equivalent summation in Fourier space
- methode to compute long-range interactions in periodic systems

### total electrostatic potential at position $\vec{r}_i$
$$
\phi(\vec{r}_i)=\frac{1}{4\pi\varepsilon _0} \sum^`_n \sum^N_{j=1}\frac{q_j}{|\vec{r}_j-\vec{r}_i + \vec{n}|}
$$
$\sum^`_n$: run over all periodic box vectors, but in the primary cell (central cell) $i\ne j$

$\sum^N_{j=1}$: all neighbored particles

$q_j$ : charge of the neighbored particle

$\vec{n}$ :  $\vec{n} = (i_x, i_y, i_z)L$ points to the origin of the periodic unit cell, with indices $i$


### Potential decomposition

Ewald summation rewrites the interacting potential as $\phi(\vec{r}_i)= \phi^{sr}(\vec{r}_i)+\phi^{lr}(\vec{r}_i)$ (I)

$\phi^{sr}$: short-range potentiol; quickly converge in real-space $\rightarrow$ real-space cutoff

$\phi^{lr}$: long-range potentiol; quickly converge in reciprocal-space $\rightarrow$ real-space cutoff; assume system is infinitv; countinuos mathematical form $\rightarrow$ Gaussian

- interaction energy/force of a single charge at position $\vec{r}$ with the field generated by all other charges

$$
\phi(\vec{r}_i) = \frac{1}{4\pi\varepsilon _0}\int d^3\vec{r}^{\prime}\frac{\rho_i(\vec{r}_i-\vec{r}^{\prime})}{|\vec{r}^{\prime}|}
$$

- rewrite charge distribution by substracing and adding smeard charge distribution $G_{\sigma}(\vec{r})$ to split this sum into $\rho^{sr}(\vec{r})$ and $\rho^{lr}(\vec{r})$ 

$$
\rho_i(\vec{r}) = q_i \delta (\vec{r}-\vec{r}_i) - G_{\sigma}(\vec{r} - \vec{r}_i) + G_{\sigma}(\vec{r} - \vec{r}_i)\\
\rho_i(\vec{r}) = \rho^{sr}(\vec{r}) + \rho^{lr}(\vec{r})
$$
    the charge ditribution of the unit cell $\rho^{uc}(\vec{r}) = q_i \delta (\vec{r}-\vec{r}_i)$ and $G_{\sigma} = \frac{1}{2\pi \sigma ^2}^{3/2}\exp(-\frac{|\vec{r}|^2}{2\sigma ^2})$
- the resulting potential is given through

$$
\phi _i(\vec{r})=\frac{q_i}{4\pi\varepsilon _0}\left[ \frac{1}{|\vec{r} - \vec{r}_i|} - \int d^3\vec{r}^{\prime}\frac{G_{\sigma}(\vec{r} - \vec{r}^{\prime})}{|\vec{r} - \vec{r}^{\prime}|}\right] + \frac{q_i}{4\pi\varepsilon _0} \int d^3\vec{r}^{\prime}\frac{G_{\sigma}(\vec{r} - \vec{r}^{\prime})}{|\vec{r} - \vec{r}^{\prime}|}
$$

    which leads to (I)

## short-range potential and interaction energy

- potential field generated by Gaussian charge distribution is evaluated by solving Poissons equation 

$$
\nabla ^2 \phi_{\sigma}(\vec{r}) = -\frac{G_{\sigma}(\vec{r})}{\epsilon _0} (*) \quad\Rightarrow\quad \phi_{\sigma}(\vec{r})\frac{q_i}{4\pi\varepsilon _0}\frac{1}{|\vec{r} - \vec{r}_i|}\text{erf}\left(\frac{|\vec{r} - \vec{r}_i|}{\sqrt{2}\sigma}\right)
$$

$$
\Rightarrow \quad  \phi^{sr}(\vec{r}_i) = \frac{q_i}{4\pi\varepsilon _0|\vec{r} - \vec{r}_i|}\left[ 1- \text{erf}\left(\frac{|\vec{r} - \vec{r}_i|}{\sqrt{2}\sigma}\right)\right]\\
\quad \quad = \frac{q_i}{4\pi\varepsilon _0|\vec{r} - \vec{r}_i|}\text{erfc}\left(\frac{|\vec{r} - \vec{r}_i|}{\sqrt{2}\sigma}\right)
$$

$\text{erf}$: error function $\text{erf}(0) = 0$ incerases within 0$\rightarrow$1
$\text{erfc}$: complementary error function $\text{erf}(0) = 1$ decay within 1$\rightarrow$0; shielded potential $\Rightarrow$ short-range

- corresponding electric energy
$$
E^{sr} = \frac{1}{8\pi\varepsilon_0}\sum_n\sum_{i=1}^N\sum_{j=1}^{N*}\frac{q_iq_j}{|\vec{r}_i-\vec{r}_j + \vec{n}|}\text{erfc}\left(\frac{|\vec{r}_i-\vec{r}_j + \vec{n}|}{\sqrt{2}\sigma}\right)
$$
    $* i\neq j$ in primary all $\Rightarrow$ not interact with itself

- total long-ranged charge density
$$
\rho^{lr}(\vec{r}) =\sum_n\sum_{j=1}^N q_j G_{\sigma}(\vec{r} - \vec{r}_i + \vec{n})\quad \quad (II)
$$

$\sum_n$ : periodic cells

- solve it by Fourier transforming the charge distribution to the recipocial space for $\phi^{lr}$ and $\rho^{lr}$, then transforming back

    - forward Fourier transforms
$$
\mathcal{F}\{ \rho^{lr}(\vec{k})\} = \int_V d^3\vec{r}\rho^{lr}(\vec{r})\exp^{-i\langle \vec{k},\vec{r}\rangle};\quad \mathcal{F}\{ \phi^{lr}(\vec{k})\} = \int_V d^3\vec{r}\phi^{lr}(\vec{r})\exp^{-i\langle \vec{k},\vec{r}\rangle}
$$

    - backward Fourier transforms
$$
\rho^{lr}(\vec{r}) = \sum_{\vec{k}}\mathcal{F}\{ \rho^{lr}(\vec{k})\}\exp^{i\langle \vec{k},\vec{r}\rangle};\quad \phi^{lr}(\vec{r}) = \sum_{\vec{k}}\mathcal{F}\{ \phi^{lr}(\vec{k})\}\exp^{i\langle \vec{k},\vec{r}\rangle}

## self-interaction

- occurs in the long-range part, because we assumed that the lattice is entirely periodic
- remove self-interaction by considering a long-range potential with $\vec r \rightarrow \vec r _i$

$$
\lim_{\vec r \rightarrow \vec r _i} \phi^{lr}_i(\vec r ) = \lim_{\vec r \rightarrow \vec r _i} \frac{q_i}{4\pi\varepsilon _0|\vec{r} - \vec{r}_i|}\text{erf}\left(\frac{|\vec{r} - \vec{r}_i|}{\sqrt{2}\sigma}\right) = \frac{q_i}{4\pi\varepsilon _0 \sigma} \sqrt{\frac{2}{\pi}}
$$

- total self-energy generated by all charges

$$
E^{self} = \frac{1}{2\varepsilon_0 (2\pi)^{3/2}}\sum^N_{i=1}q_i^2
$$
